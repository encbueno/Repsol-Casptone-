# TOURISM SEASON

In general, the low season is considered to be from November 1 to March 31 and the high season from April 1 to October 31.

However, to be more specific, some hotels and airlines consider the high season to be from November to April and from June 22 to August. The low season is considered to be from April to June 22 and the months of September and October.

Nevertheless, as always, it all depends on where you want to travel. If you are looking for a beach destination, the Canary Islands and the Balearic Islands have good weather all year round without needing to go during the high season, just like the south of the peninsula


Low season = month 11,12,1,2,3
High season= 4, 5,6,7,8 9, 10

source: ChatGPT




```python
# Function to determine season based on month (modified for index)
def determine_season(timestamp):
    month = timestamp.month  # Extract month directly from Timestamp object
    if month in [11,12,1, 2, 3]:
        return 0
    elif month in [4, 5, 6,7,8, 9, 10]:
        return 1


# Apply the function to create the new feature (using index)
merged_final3['season'] = merged_final3.index.to_series().apply(determine_season)
```
