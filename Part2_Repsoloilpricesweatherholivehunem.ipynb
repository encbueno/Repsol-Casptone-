```python
df_desem = df_desem[df_desem['year_month'].dt.year >= 2014]
```

```python
# Crear el mapa de estandarización completo
standardization_map = {
    'Andalusia': 'Andalusia',
    'Aragon': 'Aragon',
    'Asturias': 'Asturias',
    'Balearic Islands': 'Balearic Islands',
    'Canary Islands': 'Canary Islands',
    'Cantabria': 'Cantabria',
    'Castilla y Leon': 'Castile and León',
    'Castilla-La Mancha': 'Castile-La Mancha',
    'Catalonia': 'Catalonia',
    'Valencian Region': 'Valencian Region',
    'Extremadura': 'Extremadura',
    'Galicia': 'Galicia',
    'Madrid': 'Madrid',
    'Murcia': 'Murcia',
    'Navarra': 'Navarre',
    'Basque Country': 'Basque Country',
    'La Rioja': 'La Rioja',
    'Ceuta': 'Ceuta',
    'Melilla': 'Melilla'
}

# Aplicar el mapa de estandarización a df_desem
df_desem['region_standardized'] = df_desem['region'].map(standardization_map)

# Crear el mapa inverso para estandarizar df
inverse_standardization_map = {v: k for k, v in standardization_map.items()}
df['region_standardized'] = df['region'].map(inverse_standardization_map)

# Convertir year_month a datetime
df['year_month'] = pd.to_datetime(df['year_month'])
df_desem['year_month'] = pd.to_datetime(df_desem['year_month'])

# Realizar la unión usando las columnas estandarizadas y year_month
merged_final = pd.merge(df, df_desem[['year_month', 'region_standardized', 'UnemRate']],
                        left_on=['year_month', 'region_standardized'],
                        right_on=['year_month', 'region_standardized'],
                        how='left')

# Eliminar la columna temporal 'region_standardized' si ya no es necesaria
merged_final.drop(columns=['region_standardized'], inplace=True)

# Mostrar el resultado
print(merged_final.head())
```
