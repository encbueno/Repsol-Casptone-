#Vehicle landscape

```python
# Sharedrive file
file_path = '/content/drive/Shareddrives/Clandestino /Repsol/NAN/vehiclelandscap_Year_CCAA_fueltype.csv'

# second sheet to be uploaded
df_vehicles = pd.read_csv(file_path)
```

```python
df_vehicles.columns = df_vehicles.columns.str.lower()
df_vehicles.columns = df_vehicles.columns.str.replace (" ", "_")
df_vehicles.head()
```

```python
df_vehicles.region.unique()
```

```python
# Apply proposed standardization to result
df_vehicles['region'] = df_vehicles['region'].replace(standardization_map)
```

```python
# List of vehicle categories to process
vehicle_categories = ['truck', 'van', 'bus', 'car', 'motorcycle', 'tractor', 'total']

# Iterate over vehicle categories and create new columns for Gasoline and Diesel totals
for category in vehicle_categories:
    df_vehicles[f'{category}_gasoline_total'] = df_vehicles[f'{category}_gasoline']
    df_vehicles[f'{category}_diesel_total'] = df_vehicles[f'{category}_diesel']

# Create a new DataFrame with only 'Year', 'Region', and the new columns for Gasoline and Diesel totals
columns_to_keep = ['year', 'region'] + [f'{category}_gasoline_total' for category in vehicle_categories] + \
                                    [f'{category}_diesel_total' for category in vehicle_categories]
df_vehicles_cat = df_vehicles[columns_to_keep]

# Display the new DataFrame
df_vehicles_cat.head()

```

```python

# Generate a new DataFrame new_dfv with Total Gasoline Total and Total Diesel Total
df_vehicles_filtered = df_vehicles_cat[['year', 'region', 'total_gasoline_total', 'total_diesel_total']]

# Display the new DataFrame new_dfv
df_vehicles_filtered.head()
```

```python

# Rellenar los valores faltantes con ceros
df_vehicles_filtered['total_gasoline_total'] = df_vehicles_filtered['total_gasoline_total'].fillna(0)
df_vehicles_filtered['total_diesel_total'] = df_vehicles_filtered['total_diesel_total'].fillna(0)
df_vehicles_filtered.head()

```

```python
# Generar los números de los meses
months = list(range(1, 13))
```

```python
# Crear un nuevo DataFrame con los números de los meses replicados para cada año
expanded_df = df_vehicles_filtered.loc[df_vehicles_filtered.index.repeat(len(months))].reset_index(drop=True)
expanded_df['month'] = months * len(df_vehicles_filtered)
expanded_df['total_gasoline_total'] = (expanded_df['total_gasoline_total'] / 12).round().astype(int)
expanded_df['total_diesel_total'] = (expanded_df['total_diesel_total'] / 12).round().astype(int)
expanded_df.head()
```

```python
# Clean column names by stripping leading/trailing spaces
expanded_df.columns = expanded_df.columns.str.strip()
```
